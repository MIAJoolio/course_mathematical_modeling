```python
import os 
os.chdir('E://GitHubProjects/educational/mathematical_modeling/lab2/data')
import glob

import numpy as np
import pandas as pd
```

```python
files = glob.glob('*.xlsx')
data = {}
for i, file in enumerate(files):
    raw_df = pd.read_excel(file)
    if not isinstance(raw_df.iloc[0,0], str):
        raw_df.columns = [column if column in ['Unnamed: 0', 'Unnamed: 1'] else '_'.join(column.split('.')[0].split()[1:] + ['_'.join(raw_df.loc[0, column].split()[1:])]) for column in raw_df.columns]
        
    data[i] = {'raw_data':raw_df.drop(['Unnamed: 1'], axis=1).drop([0], axis=0).rename({'Unnamed: 0':'date'}, axis=1), 'original_name':file.replace('.xlsx', ''), 'indicator':file.replace('.xlsx', '').split('_')[0]}

data[0]
```

```python
# дополнительные правки
data[4]['raw_data'] = data[4]['raw_data'].rename({'6 Всего, 23320000100030200007 Миграционный прирост (убыль) населения, абсолютные данные':'Миграционный прирост (убыль) населения, абсолютные данные'}, axis=1)
data[4]['indicator'] = '23320000100030200007'

data[3]['raw_data'] = data[3]['raw_data'].rename({'31000000200040200001 ВВП России на душу населения по паритету покупательной способности (в текущих ценах)':'ВВП России на душу населения по паритету покупательной способности (в текущих ценах)'}, axis=1)
```

```python
base_df = pd.DataFrame(data[2]['raw_data'].date)
for item in [v['raw_data'] for k,v in data.items()]:
    base_df = base_df.merge(item, how='left', on = ['date'])
```

```python
base_df.head()
```

```python
rename_dict = {
                'Естественный_прирост_населения_за_год_городское_население':'growth_city',
                'Естественный_прирост_населения_за_год_сельское_население':'growth_rural',
                'Число_разводов_за_год_городское_население':'divorce_city',
                'Число_разводов_за_год_сельское_население':'divorce_rural',
                'Число_браков_за_год_городское_население':'marriage_city',
                'Число_браков_за_год_сельское_население':'marriage_rural',
                'Ожидаемая_продолжительность_жизни_при_рождении,_лет_городское_население':'LE_city',
                'Ожидаемая_продолжительность_жизни_при_рождении,_лет_сельское_население':'LE_rural',
                'ВВП России на душу населения по паритету покупательной способности (в текущих ценах)':'GDP',
                'Миграционный прирост (убыль) населения, абсолютные данные':'migration'
                }
base_df = base_df.rename(rename_dict, axis=1)
base_df
```

```python
for column in base_df.columns[1:]:
    base_df[column]=np.round(base_df[column].bfill().ffill().astype('int').values, 2)
base_df
```

```python
# создаем таргет
base_df.insert(1, 'growth_total', base_df['growth_city']+base_df['growth_rural'])
base_df.insert(2, 'divorce_total', base_df['divorce_city']+base_df['divorce_rural'])
base_df.insert(3, 'marriage_total', base_df['marriage_city']+base_df['marriage_rural'])
base_df.insert(4, 'LE_total', base_df['LE_rural']+base_df['LE_city'])
base_df
```

```python
# удаляем колонки с ростом и датой, т.к. они для дальнейшего анализа не требуются
base_df[['growth_total', 'divorce_total', 'marriage_total', 'LE_total', 'GDP', 'migration']].to_csv('final_dataset.txt', sep=';', index=False)
```
